# Book Dataset

Problem Statement

Prepare rules for the all the data sets 
1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values
2) Change the minimum length in apriori algorithm
3) Visulize the obtained rules using different plots 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv('book.csv')
book
```

# Apriori Algorithm

```python
#Association rules with 10% support
fre_itemsets=apriori(book,min_support=0.1,use_colnames=True)
fre_itemsets
```

```python
#Association rules with 70% confidence
con = association_rules(fre_itemsets,metric='lift',min_threshold=0.7)
con
```

```python
con.sort_values('lift',ascending=False)
```

```python
con[con.lift>1]
```

```python
plt.scatter(con['support'],con['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
#Association rules with 20% support
fre_itemsets2=apriori(book,min_support=0.20,use_colnames=True)
fre_itemsets2
```

```python
#Assoiciation rules with 60% confidence
con2=association_rules(fre_itemsets2,metric='lift',min_threshold=0.6)
con2
```

```python
con2.sort_values('lift',ascending=False)
```

```python
con[con.lift>1]
```

```python
plt.scatter(con2['support'],con2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
